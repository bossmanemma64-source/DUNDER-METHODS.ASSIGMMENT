ITEM ACCESS __getitem__()

```python
class Student:
    def __init__(self, name):
        self.grades = {"Math": 85, "English": 90, "Science": 78}

    def __getitem__(self, subject):
        return self.grades[subject]

s1 = Student("Ama")
print(s1["Math"])
```

ITEM ASSIGNMENT __setitem__()

```python
class Student:
    def __init__(self):
        self.grades = {}

    def __setitem__(self, subject, score):
        self.grades[subject] = score

s1 = Student()
s1["Math"] = 95
print(s1.grades)
```

MEMBERSHIP TESTING __contains__()

```python
class Library:
    def __init__(self):
        self.books = ["Python", "Java", "C++"]

    def __contains__(self, book_name):
        return book_name in self.books

lib = Library()
print("Python" in lib)
print("C#" in lib)
```

CALLABLE OBJECTS __call__()

```python
class Counter:
    def __init__(self):
        self.count = 0

    def __call__(self):
        self.count += 1
        return self.count

c = Counter()
print(c())
print(c())
```

ITERATION SUPPORT __iter__(), __next__()

```python
class Countdown:
    def __init__(self, n):
        self.n = n

    def __iter__(self):
        return self

    def __next__(self):
        if self.n > 0:
            self.n -= 1
            return self.n + 1
        raise StopIteration

for num in Countdown(3):
    print(num)
```

DESTRUCTOR __del__()

```python
class File:
    def __init__(self, name):
        self.name = name
        print(f"Opening {name}")

    def __del__(self):
        print(f"Closing {self.name}")

f = File("DUNDER METHODS")
del f
```
